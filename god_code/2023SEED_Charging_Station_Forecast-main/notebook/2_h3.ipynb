```python
# 把H3编码转换成经纬度坐标
from h3 import h3
import pandas as pd
df_stub = pd.read_csv('../../../初赛1008/训练集/stub_info.csv')
df_stub[['latitude', 'longitude']] = pd.DataFrame(df_stub['h3'].apply(lambda x: h3.h3_to_geo(x)).tolist(), 
                                                    columns=['latitude', 'longitude'])
df_stub
```

```python
# 绘制地图
import folium
def geo_map(df):
    # 拆分经纬度坐标

    # 创建地图
    m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=8)

    # 添加标记点
    
    for index, row in df.iterrows():
        
        folium.Marker(
            location=[row['latitude'], row['longitude']],
            popup=row['id_encode'],
        ).add_to(m)
    # 以指定经纬度为中心画一个圆
    for idx in [197, 416, 275, 121]:
        df_tmp = df_stub[df_stub['id_encode']==idx]
        folium.Circle(
            location=[df_tmp['latitude'].iloc[0], df_tmp['longitude'].iloc[0]],
            radius=1e3,  # 半径，单位为米
            color='blue',  # 圆的颜色
            fill=True,  # 填充圆
            fill_color='blue'  # 填充颜色
        ).add_to(m)
    # 显示地图
    return m

stub_m = geo_map(df_stub)
display(stub_m)
```
